```python
import numpy as np
import csv
import time
import pandas as pd
import matplotlib.pyplot as plt
import sys
%matplotlib inline
```

```python
f = open('yelp_output.csv')
reader = csv.reader(f, delimiter = ',')
data = []
for row in reader:
    temp = []
    for i in row:
        i.strip()
        temp.append(i)
    data.append(temp)
df = pd.DataFrame(data)
```

```python
df.columns = ['Rating', 'Review Count', 'Name', 'Cuisine', '4', '5']
```

```python
del(df['4'])
del(df['5'])
```

```python
df['Review Count'] = df['Review Count'].astype(float)
df['Rating'] = df['Rating'].astype(float)
```

```python
new_df = df.groupby(['Cuisine']).mean()
new_df
```

```python
series = pd.Series(new_df['Rating'], name='series')
series.plot(kind='pie', figsize=(14, 14))
```

From all the findings and graphs above and the analyzing from hypothesis 1, All cuisine's restaurants have really close overall rating. Thus, yelp rating system has really low possibility that has a relationship with the NYC restaurant grade.
