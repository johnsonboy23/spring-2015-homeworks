Hypothesis 1:

Will different types of cuisines treat different in the inspection? My hypothesis: There exist a relationship between type of cuisine and restaurant overall grade. In other words, I believe that western foods (i.e. American, French) are easier to get better grade than non-western foods (i.e. asian foods for example)


The data can be downloaded from https://data.cityofnewyork.us/Health/DOHMH-New-York-City-Restaurant-Inspection-Results/xx67-kt59

Handle the data and output the DF with percentage by cuisine types.

```python
def d_to_per_cuisine(input_d):
    
    output_d = input_d[['CUISINE DESCRIPTION','GRADE']].groupby('CUISINE DESCRIPTION').count()
    
    #function used to normalize grades by percentage
    perc_computation_function = \
    lambda x: x[x['GRADE'] == str(letter)].count() / x['GRADE'].count()

    for letter in input_d['GRADE'].unique():
        temp_d = input_d.groupby('CUISINE DESCRIPTION')
        temp_d = temp_d.apply(perc_computation_function)
        output_d[str(letter)] = temp_d['GRADE']    
    return output_d
```

Process the data: transform the percentage of data vs counts.

Verify the hypothesis by Chi-square Test.

Set H0: There is no relationship between cuisine type and restaurant grade

```python
d_proportions_ct = d_to_per_cuisine(d)[:]

del d_proportions_ct['nan']

# Cuisine counts
for col in d_proportions_ct.columns.tolist():
    if col == 'GRADE': continue
    d_proportions_ct[col] *= d_proportions_ct['GRADE'] 

# Set H0
total_distribution_proportion = {}
for col in d_proportions_ct.columns.tolist():
    if col == 'GRADE': continue
    total_distribution_proportion[col] = d_proportions_ct[col].sum() / d_proportions_ct['GRADE'].sum()

# Display proportions
pd.DataFrame(total_distribution_proportion, index=[0])
```

Use random sample from our cuisine datasets

```python
random_observation = np.random.randint(1, d_proportions_ct.shape[0] + 1)

# Generate expected frequency
expected = {}
rand_ob = d_proportions_ct.ix[random_observation,:].to_dict()
rand_ob_cuisine = d_proportions_ct.ix[random_observation,:].name

for key, val in total_distribution_proportion.iteritems():
    expected[key] = rand_ob['GRADE'] * val

rand_ob.pop('GRADE')

print 'Expected frequency'
print '-' * len('Expected frequency')
pprint(expected)

print ''
print 'Random observation'
print '-' * len('Random observation')
pprint(rand_ob)
```

```python
chi, p_val = stats.chisquare(rand_ob.values(), expected.values())

print ''
print 'Name: ', rand_ob_cuisine
print 'Chi-squared Test Statistic: ', chi 
print 'P-value: ', p_val
```

use α = 0.01, which means a 99% confidence interval.

If P-value is smaller than 0.01, we can reject H0. Therefore, the types of cuisine do have relationship with restaurant grade, which means some cuisines have advantages in the inspections.

Plot the violations counts. Plot in bar chart

```python
def plot_2(d_to_plot, series_name, lab):

    fig = plt.figure(figsize=(10,20))

    ax = fig.add_subplot(111)
    
    # Set color transparency (0: transparent)
    a = 0.6

    ax.xaxis.set_label_position('bottom')
    xlab = lab
    ax.set_xlabel(xlab, fontsize=20, alpha=a, ha='left')
    ax.xaxis.set_label_coords(0, 1.04)

    ax.xaxis.tick_top()

    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')

    ax.grid(False)

    ax.set_frame_on(False)

    labels = d_to_plot[series_name].value_counts().index
    labels_c = [str(labels[i]) for i, acct in enumerate(labels)]

    d_to_plot[series_name].value_counts()[::-1].plot(
    ax=ax, 
    kind='barh', 
    color=(0.7, 0.6, 0.3), 
    alpha=a, 
    edgecolor='w',
    label=labels_c, 
    fontsize=14, 
    grid=True, 
    width=.6
    )

    ax.lines[0].set_visible(False)
    
    plt.tight_layout()
    plt.show();
```
